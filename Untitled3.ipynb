```python
categorical_features = test.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_features
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(
    iterations=1000,
    learning_rate=0.03,
    depth=6,
    l2_leaf_reg=3,
    subsample=0.8,
    rsm=0.8,
    loss_function='RMSE',
    eval_metric='RMSE',
    random_seed=42,
    early_stopping_rounds=50,
    verbose=100,
    cat_features=categorical_features
)
```

```python
X_full_test=train.drop(columns=['target','target_min','target_max','target_variance','target_count'],errors='ignore')
```

```python
y_full_test=train['target']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_full_test['Place_ID'] = le.fit_transform(X_full_test['Place_ID'])
test['Date']=le.fit_transform(test['Date'])
```

```python
model.fit(X_full_test, y_full_test, cat_features=categorical_features)

```

```python
from sklearn.metrics import r2_score, mean_squared_error
y_pred = model.predict(test)
```

```python
submission = pd.DataFrame({
    'Place_ID X Date': test['Place_ID X Date'],
    'target': y_pred
})

submission.to_csv('my_submission.csv', index=False)
```
